```python
# Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
# Dataset
data = {
    "Outlook": ["Sunny","Sunny","Overcast","Rain","Rain","Rain",
                "Overcast","Sunny","Sunny","Rain","Sunny","Overcast","Overcast","Rain"],
    "Temperature": [30,32,28,25,22,24,27,29,31,23,26,28,29,21],
    "Humidity": [85,90,78,80,70,75,72,88,86,65,80,76,74,68],
    "Windy": ["No","Yes","No","No","No","Yes","Yes","No","Yes","No","Yes","Yes","No","Yes"],
    "Play": ["Don't Play","Don't Play","Play","Play","Play","Don't Play",
             "Play","Don't Play","Don't Play","Play","Play","Play","Play","Don't Play"]}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```

```python

# Label Encoding (convert categorical data to numbers)
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
```

```python
# Features and target
X = df.drop("Play", axis=1)
y = df["Play"]
```

```python
# Train Decision Tree Classifier
model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)
```

```python
# Plot Decision Tree
plt.figure(figsize=(5,5))
plot_tree(model, feature_names=X.columns, class_names=["Don't Play", "Play"], filled=True)
plt.show()
```
