```python
import pandas as pd

# Create the DataFrame
data = {
    'Age': [22, 25, 28, 35],
    'Salary': [20000, 25000, 50000, 70000],
    'Class': ['no', 'no', 'yes', 'yes']
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
df
```

```python
x = df[['Age', 'Salary']].values
y = df['Class'].map({"no" : 0 , "yes" : 1}).values
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear')
```

```python
y
```

```python
model.fit(x,y)
```

```python
model.predict([[19,15000]])
```

```python
import numpy as np
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:, 0], y)
```

```python
x[:,1]
```

```python
plt.scatter(x[:, 0], x[:, 1],c = 'red')  # w1x2 + w2x2 + b = 0 (formula of support vector classification formula),w is slope or weights,b is bias or intecept
w=model.coef_[0] #slope
b=model.intercept_[0] #intercept
x_line = np.linspace(20,40,10)
y_line = -(w[0]*x_line + b)/w[1] #formula implementation
plt.plot(x_line,y_line,color='black')
plt.scatter(45,50000,color = "black",marker = "*", s = 200)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
